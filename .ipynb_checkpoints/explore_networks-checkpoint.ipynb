```python
import networkx as nx
import os
import numpy as np
```

## Visualising graph data

```python
import xml.etree.ElementTree as ET

tree = ET.parse('gae/data/diseasome/diseasome.gexf')
root = tree.getroot()
```

```python
attributes, nodes, edges = root
```

```python
graph = nx.Graph() 

for n in nodes:
    node_name = n.attrib['id']
    node_label = n.attrib['label']
    node_type = n[0][0].attrib['value']
    d = {'label': node_label, 'type': node_type}
    if node_type == 'disease': 
        d['disease_type'] = n[0][1].attrib['value']
    graph.add_node(node_name, attr_dict=d)
    
for e in edges:
    if not 'cardinal' in e.attrib.keys(): 
        graph.add_edge(e.attrib['source'], e.attrib['target'])
```

```python
from networkx.algorithms import bipartite
disease_nodes, gene_nodes = bipartite.sets(graph)

disease_network = bipartite.weighted_projected_graph(graph, disease_nodes)
gene_network = bipartite.weighted_projected_graph(graph, gene_nodes)
```

```python
# DISEASOME
color_map = []
for node in graph:
    if graph.nodes[node]['attr_dict']['type']=='disease':
        color_map.append('red')
    else:
        color_map.append('blue')
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10, 10))
plt.legend(graph.nodes())
nx.draw(graph, node_size=12, node_color=color_map)
```

```python
import pandas as pd 

d = []
disease_nodes = []
for n in graph.nodes:
    att = graph.nodes[n]['attr_dict']
    if att['type']=='disease': 
        disease_nodes.append(n)
        d.append({'label': att['label'], 'type': att['disease_type']})

diseases = pd.DataFrame(d)
diseases = diseases.set_index(pd.Series(disease_nodes))
diseases.to_csv('gae/data/diseasome/diseases.csv')

# TODO save this dataframe, then import in 'explore_embeddings' to find class for each node,
# augument the dataframe, with the embeddings and save that. 
```

```python
disease_types = sorted(list(diseases['type'].unique()))
len(disease_types)
```

```python
from matplotlib.colors import hsv_to_rgb

def generate_distinct_colors(n, rgb=True):
    hsv = np.ones((n, 3)) * 0.8
    hsv[:, 0] = np.linspace(0, 1, n)
    return hsv_to_rgb(hsv)
    
colors = generate_distinct_colors(len(disease_types))
type_to_color = dict(zip(disease_types, colors))
```

```python
color_map = [type_to_color[disease_network.nodes[n]['attr_dict']['disease_type']] for n in disease_network.nodes]
plt.figure(figsize=(10, 10))
nx.draw(disease_network, node_size=20, node_color=color_map)
```

```python
plt.figure(figsize=(10, 10))
nx.draw(gene_network, node_size=12)
```

## Inspecting model results

```python
os.listdir('gae/data/saved/')
saved_path = 'gae/data/saved/'
```

```python
mean = np.load(saved_path+'disease_network_mean.npy')
z_sample = np.load(saved_path+'disease_network_z.npy')
y_sample = np.load(saved_path+'disease_network_y.npy')
log_std_sample = np.load(saved_path+'disease_network_log_std.npy')
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def decode(z, t):
    reconstructed_adj = sigmoid(np.matmul(z, z.T))
    reconstructed_adj = (reconstructed_adj > t) * reconstructed_adj
    return reconstructed_adj

def density(adj):
    return np.sum(adj > 0) / (adj.shape[0] * adj.shape[1])
```

```python
rec = decode(z_sample, 0.5)
density(rec)
```

```python
rec = decode(z_sample, 0.9)
density(rec)
```

```python
reconstructed = nx.from_numpy_array(rec)
plt.figure(figsize=(10, 10))
nx.draw(reconstructed, node_size=20, node_color=color_map)
```
